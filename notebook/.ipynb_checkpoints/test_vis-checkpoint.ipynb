```python
import open3d as o3d
import numpy as np
import sys,os
sys.path.append('../')

def visualize_point_cloud(xyz):
    '''
    Args:
        xyz is of shape N,3
    '''
    pcd=o3d.geometry.PointCloud()
    pcd.points=o3d.utility.Vector3dVector(xyz)
    vis=o3d.visualization.Visualizer()

    vis.create_window()
    vis.add_geometry(pcd)
    img=vis.capture_screen_float_buffer(True)
    
    return img
```

```python
from data.data_loader import PUNET_Dataset
```

```python
dataset=PUNET_Dataset()
pcd_input,gt,radius=dataset.__getitem__(0)
```

```python
pcd_xyz=pcd_input[:,0:3]
img=visualize_point_cloud(pcd_xyz)

import matplotlib.pyplot as plt
plt.imshow(img)

gt_xyz=gt[:,0:3]
gt_img=visualize_point_cloud(gt_xyz)
```
